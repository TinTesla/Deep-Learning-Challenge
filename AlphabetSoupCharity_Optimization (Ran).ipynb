## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf
from google.colab import files

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
#same
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df.drop(columns=["EIN", "NAME"], inplace=True)
#same
```

```python
# Determine the number of unique values in each column.
print(application_df.nunique())
#same
```

```python
# Look at APPLICATION_TYPE value counts for binning
print(application_df['APPLICATION_TYPE'].value_counts())
#same
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`

# decreasing cut-off from 500 to 100
application_type_counts = application_df['APPLICATION_TYPE'].value_counts()
application_types_to_replace = list(application_type_counts[application_type_counts < 100].index)

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
print(application_df['CLASSIFICATION'].value_counts())
#same
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
classification_counts = application_df['CLASSIFICATION'].value_counts()
print(classification_counts[classification_counts > 1])
#same
```

```python
# Choose a cutoff value and create a list of classifications to be replaced use the variable name `classifications_to_replace`

#decreasing cutoff from 1500 to 100
classifications_to_replace = list(classification_counts[classification_counts < 100].index)

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
application_df = pd.get_dummies(application_df)
application_df.head()
#same
```

```python
# Split our preprocessed data into our features and target arrays
X = application_df.drop(columns=["IS_SUCCESSFUL"])
y = application_df["IS_SUCCESSFUL"]

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=79)
#same
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

Optimize the Model

```python
# Define the model - deep neural net

# Decreasing nodes and added 3rd layer
input_features = len(X_train.columns)
hidden_nodes_layer1 = 20
hidden_nodes_layer2 = 10
#hidden_nodes_layer3 = 5

nn_optimized = tf.keras.models.Sequential()

# First hidden layer
nn_optimized.add(tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=input_features, activation="relu"))

# Second hidden layer - mode changed to LeakyReLU
nn_optimized.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="LeakyReLU"))

# Third hidden layer - mode changed to LeakyReLU
#nn_optimized.add(tf.keras.layers.Dense(units=hidden_nodes_layer3, activation="LeakyReLU"))

# Output layer
nn_optimized.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the optimized model
nn_optimized.summary()
```

```python
# Compile the optimized model
nn_optimized.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Adding early stop routing to counter progress plateauing
#early_stopping = tf.keras.callbacks.EarlyStopping(monitor="accuracy", patience=10, verbose=1)
```

```python
# Train the optimized model with decreased epochs
nn_optimized.fit(X_train_scaled, y_train, epochs=50) #, callbacks=[early_stopping])
```

```python
# Evaluate the optimized model
model_loss_optimized, model_accuracy_optimized = nn_optimized.evaluate(X_test_scaled, y_test, verbose=2)
print(f"Loss: {model_loss_optimized}, Accuracy: {model_accuracy_optimized}")
```

```python
# Export the optimized model to an HDF5 file
nn_optimized.save("AlphabetSoupCharity_Optimization.h5")
files.download("AlphabetSoupCharity_Optimization.h5")
```
